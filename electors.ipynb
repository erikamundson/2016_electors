```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 100
```

```python
df = pd.read_excel('excel/federalelections2016.xlsx', sheet_name=2).head(51)
```

```python
df.head()
```

```python
df['Viable Votes'] = df['Trump (R).1'] + df['Clinton (D).1']
```

```python
df.head()
```

```python
df['Electors'] = [9,3,11,6,55,9,7,3,3,29,16,4,4,20,11,6,6,8,8,4,10,11,16,10,6,10,3,5,6,4,14,5,29,15,3,18,7,7,20,4,9,3,11,38,6,3,13,12,5,10,3]
```

```python
df
```

```python
df['Clinton Electors'] = np.round(df['Electors'] * df['Clinton (D).1'] / df['Viable Votes'], 0)
```

```python
df['Trump Electors'] = np.round(df['Electors'] * df['Trump (R).1'] / df['Viable Votes'],0)
```

```python
df['Repr Electors'] = df['Electors'] - 2.0
```

```python
df.head()
```

```python
df['Clinton Win'] = (df['Clinton (D).1'] > df['Trump (R).1'])
df['Trump Win'] = (df['Trump (R).1'] > df['Clinton (D).1'])                     
```

```python
df['Clinton Electors 2'] = 0.0
df['Trump Electors 2'] = 0.0
```

```python
for i in range(len(df['STATE'])):
    df['Clinton Electors 2'][i] = np.round(df['Clinton (D).1'][i] * df['Repr Electors'][i] / df['Viable Votes'][i],0) + (2 if df['Clinton Win'][i] else 0)
    df['Trump Electors 2'][i] = np.round(df['Trump (R).1'][i] * df['Repr Electors'][i] / df['Viable Votes'][i], 0) + (2 if df['Trump Win'][i] else 0)
```

```python
summary = pd.DataFrame(columns = ['Metric', 'Trump', 'Clinton'])
```

```python
summary
```

```python
summary.loc[0] = ['State Popular Vote', 268, 270] 
summary.loc[1] = ['State Popular Vote WTA Senators', 272, 266]
summary.loc[2] = ['Viable Popular Vote * 538', 263, 275]
summary.loc[3] = ['Congressional District', 290, 248]
```

```python
summary.loc[4] = ['Actual', 304, 227]
summary.loc[5] = ['Actual w/o Faithless Electors', 305, 233]
```

```python
summary = pd.melt(summary, id_vars='Metric', var_name='Candidate', value_name='Electoral Votes')
summary
```

```python
def show_values_on_bars(axs):
    def _show_on_single_plot(ax):        
        for p in ax.patches:
            _x = p.get_x() + p.get_width() / 2
            _y = p.get_y() + p.get_height() + 5
            value = '{}'.format(int(p.get_height()))
            ax.text(_x, _y, value, ha="center", fontweight='bold') 

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _show_on_single_plot(ax)
    else:
        _show_on_single_plot(axs)
```

```python
colors = ["#DE0100", "#0015BC"]
f, ax = plt.subplots(1,1)
sns.set(rc={'figure.figsize':(20,10)})
sns.set_style('white')
sns.barplot(x='Metric', y='Electoral Votes', data=summary, hue='Candidate', palette = colors, ax=ax)
plt.xlabel('Method', fontsize=12, fontweight='semibold')
plt.ylabel('Electoral Votes', fontsize=12, fontweight='semibold')
plt.title('2016 Electoral Votes by Method, 270 to Win', fontsize=14, fontweight='bold')
plt.ylim(0, 350)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(False)
show_values_on_bars(ax)
plt.savefig('electors.png', bbox_inches='tight')
plt.show()
```
